```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
a = np.loadtxt('DataTrn.txt')
b = np.loadtxt('DataTrnLbls.txt')
c = np.loadtxt('DataTrnFrctns.txt')

data = a[:200][:]
```

```python
ave = np.average(data, axis = 0)
#標準偏差
sigma = np.std(data, axis = 0)
#もとのデータ-平均/標準偏差
data = (data-ave)/sigma
```

```python
#分散共分散行列を計算
cov_mat = np.dot(data.T, data)
#固有値分解により，固有値la と固有ベクトルvを計算
la, v = ( np.linalg.eig(cov_mat) )
```

```python
#V_2 を作成
V2 = np.array([v[0], v[1]])
#V_2 で張られる空間にデータを投影（写像の計算）
projected_data = np.dot(data, V2.T)
```

```python
for i in range(1000):
            plt.scatter(projected_data[i,0],projected_data[i,1], c='r')
plt.xlabel( 'first component' )
plt.ylabel( 'second component' )
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', fontsize=10)
plt.show()
```
